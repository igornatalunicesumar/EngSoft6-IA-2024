```python
import pandas as pd
```

```python
treino = pd.read_csv('GenerosMusicaisTreino.csv')
teste = pd.read_csv('GenerosMusicaisTeste.csv')
```

```python
#treino.shape
treino.head()
```

```python
#teste.shape
teste.head()
```

```python
X = treino.drop('GENRE', axis=1)
y = treino['GENRE']

XTeste = teste.drop('GENRE', axis=1)
yTeste = teste['GENRE']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
y_pred = clf.predict(XTeste)
```

```python
saida = pd.read_csv('Exemplo Saida.csv')
saidafinal = saida.drop('GENRE', axis=1)
```

```python
saidafinal["GENRE"] = y_pred

```

```python
saidafinal["GENRE"] = saidafinal["GENRE"].str.replace("Rock", "1:Rock")
saidafinal["GENRE"] = saidafinal["GENRE"].str.replace("Jazz", "2:Jazz")
saidafinal["GENRE"] = saidafinal["GENRE"].str.replace("Blues", "3:Blues")
saidafinal["GENRE"] = saidafinal["GENRE"].str.replace("Classical", "4:Classical")
saidafinal["GENRE"] = saidafinal["GENRE"].str.replace("Pop", "5:Pop")
saidafinal["GENRE"] = saidafinal["GENRE"].str.replace("Metal", "6:Metal")
```

```python
saidafinal.to_csv('saida.csv', index=False)
```
